```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

path_prefx = '../../../dataset/Fine-Grained-Counting-Dataset/Towards_vs_Away/'

for i in range(1, 2):
    dmap_away = np.loadtxt(
        f"{path_prefx}density_maps/away/vidf1_33_000_f00{i}.csv",
        delimiter=",",
    )
    dmap_towards = np.loadtxt(
        f"{path_prefx}density_maps/towards/vidf1_33_000_f00{i}.csv",
        delimiter=",",
    )
    original = Image.open(
        f"{path_prefx}images/vidf1_33_000_f00{i}.png"
    )

    plt.imshow(original)
    plt.imshow(dmap_away, cmap="jet", interpolation="nearest", alpha=0.5)
    plt.show()
    plt.imshow(original)
    plt.imshow(dmap_towards, cmap="jet", interpolation="nearest", alpha=0.5)
    plt.show()
```
