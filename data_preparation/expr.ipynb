```python
import scipy
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

stexample = scipy.io.loadmat(
    "../../../dataset/ShanghaiTech/part_A/train_data/ground-truth/GT_IMG_1.mat"
)
locations = np.asarray((stexample["image_info"][0])[0][0][0][0])
print(locations)

img1 = Image.open("../../../dataset/ShanghaiTech/part_A/train_data/images/IMG_1.jpg")
plt.imshow(img1)
plt.scatter(locations[:, 0], locations[:, 1], c="r", s=1)
plt.show()
# pic1_ano = anoexample["vidf1_33_003_f001.png"]
```

```python
import json

ano = open(
    "../../../dataset/Fine-Grained-Counting-Dataset/Towards_vs_Away/annotations/annotations.json"
)
anoexample = json.load(ano)

fimg1 = Image.open(
    "../../../dataset/Fine-Grained-Counting-Dataset/Towards_vs_Away/images/vidf1_33_003_f001.png"
)
pic1_ano = anoexample["vidf1_33_003_f001.png"]
towards = pic1_ano[0]
away = pic1_ano[1]

towards = np.c_[towards["x"], towards["y"]]
away = np.c_[away["x"], away["y"]]
print(towards, away)

plt.imshow(fimg1)
plt.scatter(towards[:, 0], towards[:, 1], c="r", s=1)
plt.scatter(away[:, 0], away[:, 1], c="y", s=1)
plt.show()
```
